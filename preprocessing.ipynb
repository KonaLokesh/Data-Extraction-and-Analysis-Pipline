```python
import os
import pandas as pd
from collections import Counter

# Function to perform the steps on a DataFrame and save the resulting Excel files
def process_json_file(json_file_path):
    # Read JSON file into DataFrame
    grm = pd.read_json(json_file_path)

    # Drop rows with missing values in specified columns
    grm.dropna(subset=["unique_number", "Gender"], inplace=True)

    grm.dropna(subset=["Gender","House Number"],inplace=True)

    # Clean 'Other' column
    grm["Other"] = grm["Other"].str.replace("s:", "")

    for column in grm.columns:
    grm[column] = grm[column].str.replace('!', 'i')

    # Convert 'Name' column to lowercase
    grm["Name"] = grm["Name"].str.lower()

    # Convert 'Name' column to string type
    grm['Name'] = grm["Name"].astype(str)

    # Extract last word from 'Name' column
    grm['First_Word'] = grm['Name'].str.split().str[-1]

    # Count occurrences of last words
    last_name_word1 = pd.DataFrame(Counter(grm['First_Word']).items(), columns=["word", "count"])

    # Filter last names with count greater than 1
    last_name_word = last_name_word1[last_name_word1['count'] > 1].sort_values(by="word", ascending=False)

    # Save DataFrame to Excel in the specified folders
    polling_station_name = grm["No_and_Name_of_Polling_Station"].iloc[0]  # Get the polling station name from the DataFrame

    # Save last name counts to Excel
    last_name_word1.to_excel(os.path.join(r"C:\Users\lokes\Desktop\1-10 pdfs\1-10 voters_json_files\Excel_last_name", f"{polling_station_name}.xlsx"), index=False)

    # Save DataFrame to Excel
    grm.to_excel(os.path.join(r"C:\Users\lokes\Desktop\1-10 pdfs\1-10 voters_json_files\Excel_semi_full", f"{polling_station_name}.xlsx"), index=False)

# Directory containing JSON files
json_folder = r"C:\Users\lokes\Desktop\1-10 pdfs\1-10 voters_json_files\json_auto"

# Iterate over JSON files in the folder
for filename in os.listdir(json_folder):
    if filename.endswith(".json"):
        json_file_path = os.path.join(json_folder, filename)
        process_json_file(json_file_path)
```
